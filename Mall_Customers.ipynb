```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mall=pd.read_csv('/content/Mall_Customers.csv')
mall.head()
```

```python
mall.value_counts()
mall.info()
mall.describe()
```

```python
mall.isna().sum()
mall[mall.duplicated()]
```

```python
#mall['Age'].nunique()
#mall['Age'].max()
sns.boxplot( y=mall["Age"] );
```

```python
plt.figure(figsize= (30,30))
plt.subplot(4,3,1)
plt.title('Gender Distribution')
sns.countplot(x=mall['Gender'],data=mall)



```

```python
#
plt.figure(figsize=(30,30))

plt.subplot(4,3,1)
plt.title('Age Distribution')
sns.countplot(y=mall['Age'],data=mall)


plt.subplot(4,3,2)
plt.title('Most Ages to occur')
mall.groupby(['Age'])['Gender'].count().sort_values(ascending=False).head(20).plot(kind = "barh")



plt.figure(figsize=(30,30))
plt.subplot(4,3,1)
plt.title('Male Ages Distribution')
x=mall[mall['Gender']=='Male']
sns.countplot(y=x['Age'],data=x)

plt.subplot(4,3,2)
plt.title('Most Men Ages to occur')
x.Age.value_counts().sort_values(ascending=False)[:20].plot(kind = 'barh')



plt.figure(figsize=(30,30))
plt.subplot(4,3,1)
plt.title('Female Ages Distribution')
f=mall[mall['Gender']=='Female']
sns.countplot(y=f['Age'],data=f)

plt.subplot(4,3,2)
plt.title('Most Men Ages to occur')
x.Age.value_counts().sort_values(ascending=False)[:20].plot(kind = 'barh')



```

```python
#On average Men got more incomce than women 
#On average Women Spend More than Men
plt.figure(figsize= (20,20))
plt.subplot(4,3,1)
mall.groupby(['Gender'])['Annual Income (k$)'].mean().plot(kind = "bar")

plt.subplot(4,3,2)
mall.groupby(['Gender'])['Spending Score (1-100)'].mean().plot(kind = "bar")
```

```python
#Both Gender highest incomes are between 30 and 40 years Old
#Both Gender Highest Spending rates are between 20 and 40 Years Old
plt.figure(figsize= (20,20))
x=mall[mall['Gender']=='Male']
plt.subplot(4,3,1)
plt.scatter(x['Age'],x['Annual Income (k$)'])
plt.title(' Men Age Relation with Income')
plt.xlabel("Age")
plt.ylabel("Income")


plt.subplot(4,3,2)
plt.scatter(x['Age'],x['Spending Score (1-100)'])
plt.title(' Men Age Relation with Spending Score')
plt.xlabel("Age")
plt.ylabel("Spending")




plt.figure(figsize= (20,20))
f=mall[mall['Gender']=='Female']
plt.subplot(4,3,1)
plt.scatter(f['Age'],f['Annual Income (k$)'])
plt.title(' Women Age Relation with Income')
plt.xlabel("Age")
plt.ylabel("Income")

plt.subplot(4,3,2)
plt.scatter(f['Age'],f['Spending Score (1-100)'])
plt.title(' Female Age Relation with Spending Score')
plt.xlabel("Age")
plt.ylabel("Spending")



```

```python
#Income Relation With Spending 
#Both Spending Increases when Income increases
plt.figure(figsize= (20,20))
x=mall[mall['Gender']=='Male']
plt.subplot(4,3,1)
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'])
plt.title(' Men Income vs Spending')
plt.xlabel("Income")
plt.ylabel("Spending")

plt.figure(figsize= (20,20))
f=mall[mall['Gender']=='Female']
plt.subplot(4,3,1)
plt.scatter(f['Annual Income (k$)'],f['Spending Score (1-100)'])
plt.title(' Women Income vs Spending')
plt.xlabel("Income")
plt.ylabel("Spending")
```
