```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cars=pd.read_csv('/content/cars.csv')
cars.head()
```

```python
cars['CO2'].nunique()
```

```python
cars.describe()
```

```python
cars.isna().sum()
cars.info()
cars[cars.duplicated()]
```

```python
#Distribution of cars names and moodels
plt.figure(figsize= (10,30))
plt.subplot(6,2,1)
plt.title('CarNames')
sns.countplot(y=cars['Car'],data=cars)
plt.subplot(4,2,2)
plt.title('CarModels')
sns.countplot(y=cars['Model'],data=cars)
```

```python
#Cars with Highest Volumes
plt.figure(figsize= (25,20))
plt.subplot(4,2,1)
plt.title('Cars  Highest Volumes')
max=cars[cars['Volume']>1500]
max.groupby(['Car',"Model"])["Volume"].sum().sort_values(ascending=False).plot(kind = "barh")

plt.subplot(4,2,2)
plt.title('Cars  Lowest Volumes')
min=cars[cars['Volume']<1500]
min.groupby(['Car',"Model"])["Volume"].sum().sort_values(ascending=True).plot(kind = "barh")

```

```python
#Cars With Highest Weights
plt.figure(figsize= (20,20))

plt.subplot(4,2,1)
plt.title('Cars  Highest Weights')
max=cars[cars['Weight']>1200]
max.groupby(['Car',"Model"])["Weight"].sum().sort_values(ascending=False).plot(kind = "barh")

plt.subplot(4,2,2)
plt.title('Cars  Lowest Weights')
min=cars[cars['Weight']<1200]
min.groupby(['Car',"Model"])["Weight"].sum().sort_values(ascending=True).plot(kind = "barh")
```

```python
#Cars With Highest CO2

plt.figure(figsize= (20,20))

plt.subplot(4,2,1)
plt.title('Cars  Highest CO2')
max=cars[cars['CO2']>100]
max.groupby(['Car',"Model"])["CO2"].sum().sort_values(ascending=False).plot(kind = "barh")

plt.subplot(4,2,2)
plt.title('Cars  Lowest Weights')
min=cars[cars['CO2']<100]
min.groupby(['Car',"Model"])["CO2"].sum().sort_values(ascending=True).plot(kind = "barh")
```

```python
#Volume and Weight Relations
plt.figure(figsize= (20,20))

plt.subplot(4,3,1)
plt.scatter(cars['Volume'],cars['Weight'])
plt.title('Volume and Weight Relation')
plt.xlabel("Volume")
plt.ylabel("Weight")

plt.subplot(4,3,2)
plt.title('Volume and CO2 Relation')
plt.xlabel("Volume")
plt.ylabel("CO2")
plt.scatter(cars['Volume'],cars['CO2'])

plt.subplot(4,3,3)
plt.title('Weight and Co2 Relation')
plt.scatter(cars['Weight'],cars['CO2'])
plt.xlabel("Weight")
plt.ylabel("CO2")
```

```python
#HeatMap to show relations between Volume, Weight , and CO2
cormat = cars.corr()
round(cormat,2)
cmap = sns.diverging_palette(220, 15, as_cmap=True)
sns.heatmap(
    cormat,          # The data to plot
    cmap=cmap,     # What colors to plot the heatmap as
    annot=True,    # Should the values be plotted in the cells?
    square=True)  # Force cells to be square
```
