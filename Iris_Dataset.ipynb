```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris=pd.read_csv("/content/Iris.csv")
iris.head()
```

```python
#@title cross tab
pd.crosstab(iris[''])
```

```python
iris.info()
```

```python
iris.nunique()
```

```python
#@title No Null Values
iris.isna().sum()
```

```python
#@title No Duplication
iris.duplicated().sum()
```

```python
#@title Histogram for all Species Studied
numeric=iris.drop(['Id','Species'],axis=1)
plt.figure(figsize=(17,13))
for i in list(enumerate(numeric.columns)):
    plt.subplot(3, 4, i[0]+1)
    sns.histplot(data = numeric[i[1]], kde=True)  # Histogram with KDE line
plt.suptitle("All Species Distribution",y=0.90)
```

**Sepal Width Shows Normal Distribution Relation as most values occur within 2.5 and 3.5**

```python
#@title Count of Each Species
plt.figure(figsize= (5,5))

sns.countplot(x=iris['Species'],data=iris)
```

**We have balanced data for all classes**

```python
#@title Correlation Between Numeric Features for all species
plt.figure(figsize=(5,5))
plt.title("Correlation Between All Columns")
cor = numeric.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)

fig=plt.figure(figsize=(10,10))

counter=1
for row in numeric.columns:
  for column in numeric.columns:
    plt.subplot(4,4,counter)
    plt.xlabel(row)
    plt.ylabel(column)
    plt.scatter(iris[row],iris[column])
    counter=counter+1
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

plt.suptitle("Relation Between each 2 Features ",y=0.92)

# sns.pairplot(iris,hue='Species')

```

```python
#@title Unique Values in Species
iris['Species'].unique()
setosa=iris[iris['Species']=='Iris-setosa']
versicolor=iris[iris['Species']=='Iris-versicolor']
virginica=iris[iris['Species']=='Iris-virginica']
setosa=setosa.drop(['Id','Species'],axis=1)
versicolor=versicolor.drop(['Id','Species'],axis=1)
virginica=virginica.drop(['Id','Species'],axis=1)
```
